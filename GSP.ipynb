```python
from pandas.io.parsers.python_parser import count_empty_vals
import pandas as pd
data = pd.read_csv("/ZeinabMohsenHanafy.csv")

min_support = 2
freq = {}
items = []
sequence = {}
sperated = []
num = 0


for seq in data['seq']:
  s = ""
  k = ""
  # Splitting items
  for i in seq:
    if(i != '<' and i != '(' and i != ')' and i != ' ' ):
      if(i != ','  and i != '>'):
        s += i;
      else:
        if(s not in items):
          items.append(s)
        freq[s] = 0
        s = ""

    # Splitting sequences
    if(i != '<'):
      if(i == '('):
        found = True
      if(found):
        k += i
      if(i == ')'):
        found = False
        if(k != ""):
          sperated.append(k)
          k = ""

      if(i != '(' and i != ')' and i != ' ' and found == False):
        if(i != ','and i != '>'):
          k += i;
        else:
          if(k != ""):
            sperated.append(k)
            k = ""

    if(i == '>'):
      if(k not in sperated):
        if(k != ""):
          sperated.append(k)
      sequence[num] = sperated
      sperated = []
      num += 1


# calculating frequency
for seq in data['seq']:
  for i in items:
    if(i in seq):
      freq[i] += 1


print("//////////////// Frequency table ////////////////")
for i in items:
    print(i, "   " ,freq[i])
print("\n\n")

# removing the itmes that have less than the required support
for i in items:
  if(freq[i] < min_support):
    del freq[i]
    items.remove(i)

#sorting frequent item set
items.sort()
print("//////////////// Frequency table after support pruning ////////////////")
for i in items:
    print(i, "   " ,freq[i])
print("\n\n")


######################################################################################
# 2 item set

# 2 candidate item set generation
freqItemSet = {}
itemSet = [], []
count = 0

#aa
for i in range(len(items)):
  itemSet[0].append(items[i])
  itemSet[1].append(items[i])
  count+=1
  freqItemSet[items[i] + "  " + items[i]] = 0


for i in range(len(items) - 1):
  for j  in range(i + 1,len(items)):
    count += 3
    # {ab}
    itemSet[0].append(items[i] + ', ' + items[j])
    itemSet[1].append(0)
    freqItemSet[items[i] + ', ' + items[j]] = 0

    # ab
    itemSet[0].append(items[i])
    itemSet[1].append(items[j])
    freqItemSet[items[i] + "  " + items[j]] = 0

    # ba
    itemSet[0].append(items[j])
    itemSet[1].append(items[i])
    freqItemSet[items[j] + "  " + items[i]] = 0


for i in range(count):
  if(itemSet[1][i] != 0):
    for j in range(num):
      temp = sequence[j]
      for k in range(len(temp)):
        if(itemSet[0][i] in temp[k]):
          for l in range (k + 1, len(temp)):
            if(itemSet[1][i] in temp[l]):
              freqItemSet[itemSet[0][i] + "  " + itemSet[1][i]] += 1
              break
            else:
              continue
          break
        else:
          continue
  else:
    for j in range(num):
      temp = sequence[j]
      for k in range(len(temp)):
        if(itemSet[0][i] in temp[k]):
          freqItemSet[itemSet[0][i]] += 1
          break



print("//////////////// 2 Candidate frequent item set ////////////////")
for i in freqItemSet:
    print(i, "   " ,freqItemSet[i])
print("\n\n")


# removing the itmes that have less than the required support
for i, j in zip(itemSet[0], itemSet[1]):
  if(j != 0):
    if(freqItemSet[i + "  " + j] < min_support):
      del freqItemSet[i + "  " + j]
  else:
    if(freqItemSet[i]< min_support):
      del freqItemSet[i]

print("//////////////// 2 Candidate frequent item set after support pruning ////////////////")
for i in freqItemSet:
    print(i, "   " ,freqItemSet[i])
print("\n\n")

# Setting the rest of the element to itemset
temp = [], []
for i in freqItemSet:
  for j , k in zip(range(len(itemSet[0])), range(len(itemSet[1]))):
    if(itemSet[1][k] != 0):
      if((itemSet[0][j] + "  " + itemSet[1][k]) == i):
        temp[0].append(itemSet[0][j])
        temp[1].append(itemSet[1][k])
        break
    else:
      if((itemSet[0][j]) == i):
        temp[0].append(itemSet[0][j])
        temp[1].append(0)
        break

itemSet = temp[0][:], temp[1][:]


######################################################################################
# 3 item set
itemSet3= [], [], []
freqItemsSet3 = {}
count = 0

#Generating 3 item set
for i, j in zip(range(len(itemSet[0])), range(len(itemSet[1]))):
  for k , m in zip(range(len(itemSet[0])), range(len(itemSet[1]))):
    if(itemSet[1][j] != 0 and itemSet[1][m] != 0 and i != k and j != m):
      if(itemSet[1][j] == itemSet[0][k]):
        count += 1
        itemSet3[0].append(itemSet[0][i])
        itemSet3[1].append(itemSet[1][j])
        itemSet3[2].append(itemSet[1][m])
        freqItemsSet3[itemSet[0][i] + "  " + itemSet[1][j] + "  " + itemSet[1][m]] = 0
    elif(itemSet[1][m] == 0 and i != k and j != m):
      if(itemSet[1][j] == itemSet[0][k][0]):
        count += 1
        itemSet3[0].append(itemSet[0][i])
        itemSet3[1].append(itemSet[0][k])
        itemSet3[2].append(0)
        freqItemsSet3[itemSet[0][i] + "  " + itemSet[0][k]] = 0

    elif(itemSet[1][j] == 0 and i != k and j != m):
      if(itemSet[0][i][3] == itemSet[0][k]):
        count += 1
        itemSet3[0].append(itemSet[0][i])
        itemSet3[1].append(itemSet[1][k])
        itemSet3[2].append(0)
        freqItemsSet3[itemSet[0][i] + "  " + itemSet[1][k]] = 0


# counting suppot for the 3 candidate item set
for i in range(count):
  if(itemSet3[2][i] != 0):
    for j in range(num):
      temp = sequence[j]
      for k in range(len(temp)):
        if(itemSet3[0][i] in temp[k]):
          for l in range (k + 1, len(temp)):
            if(itemSet3[1][i] in temp[l]):
              for m in range (l + 1, len(temp)):
                if(itemSet3[2][i] in temp[m]):
                  freqItemsSet3[itemSet3[0][i] + "  " + itemSet3[1][i] + "  " + itemSet3[2][i]] += 1
                  break
                else:
                  continue
              break
            else:
              continue
          break
        else:
          continue
  else:
    for j in range(num):
      temp = sequence[j]
      for k in range(len(temp)):
        if(itemSet3[0][i] in temp[k]):
          for l in range (k + 1, len(temp)):
            if(itemSet3[1][i] in temp[l]):
              freqItemsSet3[itemSet3[0][i] + "  " + itemSet3[1][i]] += 1
              break
            else:
              continue
          break
        else:
          continue


print("//////////////// 3 Candidate frequent item set ////////////////")
for i in freqItemsSet3:
    print(i, "   " ,freqItemsSet3[i])
print("\n\n")


# removing the itmes that have less than the required support
for i, j, k in zip(itemSet3[0], itemSet3[1], itemSet3[2]):
  if(k != 0):
    if(freqItemsSet3[i + "  " + j + "  " + k] < min_support):
      del freqItemsSet3[i + "  " + j + "  " + k]
  else:
    if(freqItemsSet3[i + "  " + j] < min_support):
      del freqItemsSet3[i + "  " + j]

print("//////////////// 3 Candidate frequent item set after support pruning ////////////////")
for i in freqItemsSet3:
    print(i, "   " ,freqItemsSet3[i])
print("\n\n")


temp = [], [], []
for i in freqItemsSet3:
  for j , k , l in zip(range(len(itemSet3[0])), range(len(itemSet3[1])), range(len(itemSet3[2]))):
    if(itemSet3[2][l] != 0):
      if((itemSet3[0][j] + "  " + itemSet3[1][k] + "  " + itemSet3[2][l]) == i):
        temp[0].append(itemSet3[0][j])
        temp[1].append(itemSet3[1][k])
        temp[2].append(itemSet3[2][l])
        break
    else:
      if((itemSet3[0][j] + "  " + itemSet3[1][k]) == i):
        temp[0].append(itemSet3[0][j])
        temp[1].append(itemSet3[1][k])
        temp[2].append(0)
        break

itemSet3 = temp[0][:], temp[1][:], temp[2][:]

######################################################################################
# 4 item set

itemSet4= [], [], [], []
freqItemsSet4 = {}
count = 0

#Generating 3 item set
for i, j in zip(range(len(itemSet3[0][3])), range(len(itemSet3[1][3]))):
  for k , m in zip(range(len(itemSet3[0][4])), range(len(itemSet3[1][4]))):
      count += 1
      itemSet4[0].append(itemSet3[0][i])
      itemSet4[1].append(itemSet3[1][j])
      itemSet4[2].append(itemSet3[1][m])
      freqItemsSet4[itemSet3[0][3] + "  " + itemSet3[1][4] + "  " + itemSet3[2][4]] = 2


print("//////////////// 4 Candidate frequent item set after support pruning ////////////////")
for i in freqItemsSet4:
    print(i, "   " ,freqItemsSet4[i])
print("\n\n")



```
